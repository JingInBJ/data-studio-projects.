```python
import requests
import urllib.request
import re
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rc
```

```python
ratio = pd.read_csv('1-M-GDP-ratio.csv')
```

```python
ratio
```

```python
rationew=ratio.transpose()
rationew
```

```python
rationew.columns = rationew.iloc[0]
```

```python
rationew
```

```python
df_ratio = rationew.drop(rationew.index[0])
```

```python
df_ratio
```

```python
df_ratio.info()
```

```python
df_ratio['US']=df_ratio['US'].astype('float').round(1)
df_ratio['Canada']=df_ratio['Canada'].astype('float').round(1)
df_ratio['France']=df_ratio['France'].astype('float').round(1)
df_ratio['Germany']=df_ratio['Germany'].astype('float').round(1)
df_ratio['Italy']=df_ratio['Italy'].astype('float').round(1)
df_ratio['Japan']=df_ratio['Japan'].astype('float').round(1)
df_ratio['UK']=df_ratio['UK'].astype('float').round(1)
#应该有更简便的方法，一次性清理所有的国家列
```

```python
df_ratio
```

```python
fig, ax = plt.subplots(figsize=(20, 10))


plt.title("G7's Manufacturing output(% of GDP)1997-2018")
ax.title.set_fontsize(20)

ax.plot(df_ratio.index,
        df_ratio["Japan"],
        color = "#0c2c84",
        linewidth=5.0)

ax.plot(df_ratio.index,
        df_ratio["Germany"],
        color = "#225ea8",
        linewidth=5.0)

ax.plot(df_ratio.index,
        df_ratio["Italy"],
        color = "#1d91c0",
        linewidth=5.0)

ax.plot(df_ratio.index,
        df_ratio["Canada"],
        color = "#41b6c4",
        linewidth=5.0)

ax.plot(df_ratio.index,
        df_ratio["US"],
        color = "red",
        linewidth=5.0)

ax.plot(df_ratio.index,
        df_ratio["UK"],
        color = "#7fcdbb",
        linewidth=5.0)

ax.plot(df_ratio.index,
        df_ratio["France"],
        color = "#c7e9b4",
        linewidth=5.0)

#labels做list的括号里的国家顺序决定了图例的顺序，并不会自动匹配颜色顺序
labels = list(['Japan','Germany','Italy','Canada','US','UK','France'])  
plt.legend(labels, loc='bottom left')

plt.savefig("G7.pdf", transparent=False)
```

```python
df_jobs = pd.read_csv('2-jobs.csv')
#employment data source: https://www.bls.gov/charts/employment-situation/employment-levels-by-industry.htm
```

```python
df_jobs.head()
```

```python
df_jobs.info()
```

```python
df_jobs['Total nonfarm']=df_jobs['Total nonfarm'].str.replace(',', '')
#注意，必须要先.str，再用replace才能成功，不然直接使用replace虽然不会报错，但是无法删除逗号
```

```python
df_jobs.head()
```

```python
df_jobs['Total nonfarm']=df_jobs['Total nonfarm'].astype(int)
df_jobs.info()
```

```python
#同样的方法处理Manufacturing 
df_jobs['Manufacturing']=df_jobs['Manufacturing'].str.replace(',', '')
```

```python
df_jobs['Manufacturing']=df_jobs['Manufacturing'].astype(int)
```

```python
df_jobs.info()
```

```python
df_jobs['Mratio']=df_jobs['Manufacturing']/df_jobs['Total nonfarm']*100
df_jobs['Others']=df_jobs['Total nonfarm']-df_jobs['Manufacturing']
df_jobs['Oratio']=df_jobs['Others']/df_jobs['Total nonfarm']*100
```

```python
df_jobs.head()
```

```python
df_jobs['Month']=pd.to_datetime(df_jobs['Month'])
#还是需要处理时间的，希望时间能够只显示年和月
```

```python
import os
```

```python
#这一步能把YMD修改为了YM，但是int格式,无法直接应用于表格

df_jobs['Time'] = df_jobs['Month'].map(lambda x: 100*x.year + x.month)
```

```python
df_jobs.head()
```

```python
df_jobs.info()
```

```python
# plot
f, ax = plt.subplots(figsize=(60,30))

barWidth = 10
names = df_jobs['Month']

# Create Manufacturing Bars
plt.bar(df_jobs['Month'], df_jobs['Mratio'], color='#2b8cbe',width=barWidth)

# Create Others Bars
plt.bar(df_jobs['Month'], df_jobs['Oratio'], bottom=df_jobs['Mratio'], color='#f9bc86', width=barWidth)



```
